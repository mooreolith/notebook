```javascript
/*
In this experiment, I measure 6 looping expressions and functions: 

1. a regular for loop
2. the reduce function
3. the for-of loop
4. the for-in loop
5. the forEach function
6. the map function

The different constructs are used to add up an array filled with n 1s, n being
a variable defined at the beginning of the source code. 
*/
```

```javascript
const n = 1e5;
const array = (new Array(n)).fill(1);

let a = 0, starta = performance.now();
for(let i=0; i<n; i++) a += array[i];
const ta = performance.now() - starta;

let b, startb = performance.now();
b = array.reduce((b, e) => b + e, 0);
const tb = performance.now() - startb;

let c = 0, startc = performance.now()
for(let e of array) c += e;
const tc = performance.now() - startc;

let d = 0, startd = performance.now();
for(let e in array) d += e;
const td = performance.now() - startd;

let f = 0, startf = performance.now();
array.forEach(e => f += e);
const tf = performance.now() - startf;

let g = 0, startg = performance.now();
array.map(e => g += e);
const tg = performance.now() - startg;

let i=0, h = 0; starth = performance.now();
while(i++ < array.length) h += array[i];
const th = performance.now() - starth;

const times = [
  {t: ta, name: 'for'},
  {t: tb, name: 'reduce'},
  {t: tc, name: 'for-of'},
  {t: td, name: 'for-in'},
  {t: tf, name: 'forEach'},
  {t: tg, name: 'map'},
  {t: th, name: 'while'}
].sort((a, b) => a.t - b.t)

console.log(`Addition of ${n.toExponential()} 1s:`)
for(let m of times){
  console.log(`${m.t} ms - ${m.name}`)
}
```

```javascript
/*
A typical run lists either forEach or reduce as the most time efficient ways to
add all n 1s in the array. 

This is a sample output of the program above: 

Addition of 1e+5 1s:
1 ms - reduce
2 ms - forEach
2 ms - map
8 ms - for
9 ms - for-of
13 ms - for-in

The results are sorted by time, and reduce and forEach more or less take turns
for the spot of fastest array access. 
*/
```
