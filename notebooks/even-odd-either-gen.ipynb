```javascript
/*
  Returns two generators in an object, .odd and .even, which return whether either has been called an odd
  or even number of times.

    Usage:

  var is = isOddEvenGens();
  console.log(is.odd.next().value); // prints true
  console.log(is.odd.next().value); // prints false
  console.log(is.even.next().value); // prints false
  console.log(is.even.next().value); // prints true
*/
function isOddEvenGens(){
  var isOdd = true;

  function* oddGen(){
    while(true){
      yield isOdd;
      isOdd = !isOdd;
    }
  }

  function* evenGen(){
    while(true){
      yield isOdd;
      isOdd = !isOdd;
    }
  }
  
  return {
    odd: oddGen(),
    even: evenGen()
  }
}
```

```javascript
var is = isOddEvenGens();
console.log(is.odd.next().value);
console.log(is.odd.next().value);
console.log(is.even.next().value);
console.log(is.even.next().value);
```
