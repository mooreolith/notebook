### The Code Cell
The Code Cells' contents run like the inside of an async javascript function. You can use a `return "xyz";` statement to print something to the cell's output field, or manipulate the `output` value property directly, via `output.value = "xyz";`. The output variable is a `HTMLOutputElement`, so you can style it to your liking.

```javascript
output.style.fontSize = "24px";
output.style.border = "1px solid orangered";
output.value = "Hello, World!"
```

#### Code Cell Return Values
Code cells can have different types of return values:
* HTMLElement: Adds the element to output's innerHTML, the output is the element
* string: displays the string
* number: displays the number
* boolean: displays the strings "true" or "false"
* object: displays the JSON.stringified object
* array: displays the JSON.stringified array
* null: displays the string "null"
* undefined: displays nothing

```javascript
const p = document.createElement('p');
p.innerText = "Goodbye, World!";
p.style.border = "1px solid orange";

// adds the element to the output.innerHTMl
return p;
```

#### Global Variables
To persist a variable across multiple cells, declare it as a global variable, ie without const, let, or var in front of it. The code below defines two variables, answer and thoughts. A cell later, answer is returned, while a cell after that, thoughts cannot be accessed from the cell it's supposed to be returned from. 

```javascript
// global variable, accessible to cells run after this one
answer = 42;

// local variable, inaccessible to other cells
const thoughts = '1 cent'
```

```javascript
// This works, answer is a `global` variable from the cell above.
return answer;
```

```javascript
// This doesn't work, `thoughts` is defined as a local variable.
return thoughts;
```
