# Dynamic Import Example: Vega Embed

To demonstrate how external libraries can be loaded (respecting CORS), we import the [vega-embed](https://vega.github.io/vega-embed/) visualization library from [esm.sh](https://esm.sh) using a [dynamic import](https://developer.mozilla.org/en-US/docs/Web/JavaScript/Reference/Operators/import). 

After importing, we define a url for the sample specification. 

Lastly, we call embed and pass it the cell's `output` variable and the sample specification URL. 

```javascript
const { default: embed } = await import('https://esm.sh/vega-embed');
const spec = 'https://raw.githubusercontent.com/vega/vega/master/docs/examples/bar-chart.vg.json';
embed(output, spec);
```
